## Tutorial No.13
Working with Frame or iFrame

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome(executable_path='/home/aadya/0AutomationPython/chromedriver')
driver.implicitly_wait(1000) #Wait maximum of 1000 seconds for each lines of codes
driver.get("https://the-internet.herokuapp.com/nested_frames")
level1=driver.find_elements_by_tag_name('frame')
print("Total frames in Lavel1: ", len(level1))
for i in range(len(level1)):
    try:
        driver.switch_to.frame(level1[i])
        level2=driver.find_elements_by_tag_name('frame')
        print("Total frames in Lavel2: ", len(level2))
        for j in range(len(level2)):
            driver.switch_to.frame(level2[j])
            print(driver.find_element_by_xpath("/html/body").text)
            driver.switch_to_default_content()
            driver.switch_to.frame(level1[i])
    except:
        driver.switch_to_default_content()
        driver.switch_to.frame(level1[i])
        print(driver.find_element_by_xpath("/html/body").text)
driver.quit()
```
